```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset={
    'Service_Type':['Local','Express','Night','Weekend','Express','Local','Night','Weekend','Local','Express'],
    'Travel_Distance':[5,40,80,10,55,3,90,None,2,100],
    'Travel_Time':[15,60,None,25,85,10,140,35,8,170],
    'Day_of_Week':['Weekday','Weekday','Weekend','Weekend','Weekday','Weekday','Weekend','Weekend','Weekday','Weekday'],
    'Passenger_Age':[25,19,68,34,22,45,72,17,29,55],
    'Ticket_Type':['Regular','Student','Senior','Regular','Student','Regular','Senior','Student','Regular','Regular']
    }
```

```python
data1=pd.DataFrame(dataset)
data1
```

```python
data1['Travel_Distance']=data1['Travel_Distance'].fillna(data1['Travel_Distance'].mean())
data1['Travel_Time']=data1['Travel_Time'].fillna(data1['Travel_Time'].mean())

data1
```

```python
def normalize(data,col):
    data[col]=(data[col]-data[col].mean())/data[col].std()
```

```python
normalize(data1,'Travel_Distance')
normalize(data1,'Travel_Time')
normalize(data1,'Passenger_Age')
data1
```

```python
encoder=LabelEncoder()
```

```python
data1['Service_Type']=encoder.fit_transform(data1['Service_Type'])
data1['Day_of_Week']=encoder.fit_transform(data1['Day_of_Week'])
data1['Ticket_Type']=encoder.fit_transform(data1['Ticket_Type'])
data1
```

```python
x=data1.iloc[:,1:]
y=data1.iloc[:,0]
```

```python
model=DecisionTreeClassifier()
rfe=RFE(model,n_features_to_select=2)
fit=rfe.fit(x,y)
fit.get_feature_names_out()
```

```python
data2=pd.DataFrame(dataset)
```

```python
data2['Travel_Distance']=data2['Travel_Distance'].fillna(data2['Travel_Distance'].mean())
data2['Travel_Time']=data2['Travel_Time'].fillna(data2['Travel_Time'].mean())

data2
```

```python
data2['Travel_Distance']=pd.cut(x=data2['Travel_Distance'],bins=[0,33,66,100],labels=['short','medium','long'])
```

```python
data2
```

```python
transactions=[[i,j,k] for i,j,k in zip(data2['Travel_Distance'],data2['Day_of_Week'],data2['Service_Type'])]
transactions
```

```python
tenc=TransactionEncoder()
enc_data=tenc.fit(transactions).transform(transactions)
enc_df=pd.DataFrame(enc_data,columns=tenc.columns_)
enc_df
```

```python
freq_itemset=apriori(enc_df,min_support=0.1,use_colnames=True)
freq_itemset
```

```python
rules=association_rules(freq_itemset,metric='confidence',min_threshold=0.3)
rules
```

```python
filtered_rules=rules[rules['consequents'].apply(lambda x:all( item in ['Local','Express','Night','Weekend'] for item in x))]
filtered_rules
```
